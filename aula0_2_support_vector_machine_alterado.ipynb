Neste notebook, você experimentará com um tipo de Support Vector Machine Linear.

## Preparação dos dados

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn import datasets

# Preparar o dataset
iris = datasets.load_iris()
X = iris.data; y = iris.target

X = X[y != 1] ; y = y[y != 1] # versicolor
y = torch.tensor(y, dtype=torch.float32)
y[y == 0] = -1  # SVM espera rótulos em {-1, +1}

X = torch.tensor(X, dtype=torch.float32) # Tensor é um tipo especial que suporta muitas dimensões
```

A nossa Support Vector Machine é basicamente um hiperplano definido por *w* e *b* que melhor separa as classes.

```python
# Definir modelo da Support Vector Machine como uma camada linear
n_features = X.shape[1]
linear = nn.Linear(n_features, 1)  # contém w (weight) e b (bias)

# === Hiperparâmetros ===
learning_rate = 0.01
epochs = 300
optimizer = optim.Adam(linear.parameters(), lr=learning_rate)
```

## Execução do treinamento

```python
for epoch in range(epochs):
    optimizer.zero_grad()

    y_pred = linear(X)  # Modelo SVM (um hiperplano que depende de w e b)

    # Hinge loss: max(0, 1 - y_i * (w^T x_i + b))
    perda_de_classificacao = torch.clamp(1 - y.view(-1, 1) * y_pred, min=0).mean()

    # Termo de regularização
    perda_de_distancia_entre_classes = 0.5 * torch.sum(linear.weight ** 2) # 2/||w|| é a distância que queremos que seja a maior possível

    # Função objetivo tradicional: minimizar reg + C * hinge
    loss = perda_de_distancia_entre_classes + perda_de_classificacao

    loss.backward()
    optimizer.step()

    if (epoch + 1) % 100 == 0:
        print(f"Epoch {epoch+1}/{epochs}, Loss={loss.item():.4f}")
```

## Avaliação do modelo


```python
# Avaliação do modelo
with torch.no_grad():
    y_pred = linear(X)
    y_class = torch.sign(y_pred).squeeze()  # converte saída para {-1, +1}

acertos = (y_class == y).sum().item()
total = len(y)
acuracia = acertos / total * 100
print(f"Acurácia: {acertos}/{total} = {acuracia:.1f}%")
```
